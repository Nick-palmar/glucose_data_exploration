```python
import pandas as pd
df = pd.read_pickle('march_df')
```

### get lows during night time
- Lows during sleep can be dangerous if not treated
- examine any trends in night time lows 

```python
only_night_lows = df.loc[(df['mmol/L'] < 5) & ((df['hours_time'] > 23) | (df['hours_time'] < 8))].reset_index()
only_night_lows
```

```python
import numpy as np
import seaborn as sns
# draw a histogram to see lows by day of week - overshoot by 2 to reach the 31st
bounds = np.arange(1, 33)
hist, edges = np.histogram(only_night_lows['day'], bounds)
rel_freq = hist/float(hist.sum())
sns.set(rc={'figure.figsize': (10, 6)})
march_lows = sns.barplot(x=bounds[:-1], y=rel_freq)
march_lows.set_xlabel('Day')
march_lows.set_ylabel('% Lows')
march_lows.set_title('March Night Lows')
```

```python
# graph march lows by time
mar_weekday_low = sns.histplot(x=only_night_lows['hours_time'], bins=40, hue=only_night_lows['weekday'].map({0: 'Mon', 2: 'Wed', 3: 'Thurs', 4: 'Fri'}))
mar_weekday_low.set_title('Weekday vs Time of Low')
```

```python
mar_weekday_low.figure.savefig('graphs/march_weekday_time_low.png')
```

```python
mar_day_num_low = sns.histplot(x=only_night_lows['hours_time'], bins=40, hue=only_night_lows['day'], palette='colorblind')
mar_day_num_low.set_title('March Day Number vs Time')
```

```python
mar_day_num_low.figure.savefig('graphs/march_day_num_time_low.png')
```

```python
# get days before when the bs was low
before_n_low = (only_night_lows['day'] - 1).unique()
before_n_low
```

```python
before_n_low_df = df.loc[df['day'].isin((only_night_lows['day'] - 1).unique())].reset_index()
before_n_low_df
```

```python
before_n_low_df.columns
```

```python
exercise_oh_cols = ['Elliptical', 'Running', 'Walking',
       'core training', 'functional strength training',
       'high intensity interval training', 'other', 'soccer', 'yoga']
exercise_pre_n_low_df = before_n_low_df.loc[(before_n_low_df[exercise_oh_cols] == 1).apply(lambda row: row.sum(), axis=1).gt(0)]
exercise_pre_n_low_df
```

```python
exercise_pre_n_low_df.groupby('day')['exercise (mins)'].sum()
```

```python
total_daily_exercise = df.groupby('day')['exercise (mins)'].sum()
median_exercise = total_daily_exercise.median()
median_exercise
```

```python
# find out what exercise was done on each day before lows
# !pip install plotly
import plotly.graph_objects as go

# use a list comprehnsion to get all the bar graphs
data = [go.Bar(name=col, x=exercise_pre_n_low_df['day'], y=exercise_pre_n_low_df[col]) for col in exercise_oh_cols]

fig = go.Figure(data=data)

fig.update_layout(barmode='group',
    width=1000,
    height=400,
    autosize=False,
    title='Exercise Before Night Lows',
    xaxis_title='Day of Month (March)',
    yaxis_title='Amount of Exercise (1 block = 5min)')
fig.show()
```

```python
# find the mean and median exercise on low days
mean_low_exercise = (exercise_pre_n_low_df.groupby('day')[exercise_oh_cols].sum()).mean()
median_low_exercise = (exercise_pre_n_low_df.groupby('day')[exercise_oh_cols].sum()).median()
print('Mean Exercise Before Bad Nights:\n' + str(mean_low_exercise))
print('Median Exercise Before Bad Nights:\n' +  str(median_low_exercise))
```

```python
low_days = [2, 7, 14, 24, 25]
non_night_low_df = df.loc[~df['day'].isin(low_days)]
non_night_low_df.loc[non_night_low_df['day'].isin(low_days)]
```

```python
non_night_low_df.head()
```

```python
# use a list comprehnsion to get all the bar graphs
data = [go.Bar(name=col, x=non_night_low_df['day'], y=non_night_low_df[col]) for col in exercise_oh_cols]

fig = go.Figure(data=data)

fig.update_layout(barmode='group',
    width=1000,
    height=400,
    autosize=False,
    title='Exercise Before Good Nights',
    xaxis_title='Day of Month (March)',
    yaxis_title='Amount of Exercise (1 block = 5min)')
fig.show()
```

```python
# find the mean and median exercise on non-low days
mean_non_low_exercise = (non_night_low_df.groupby('day')[exercise_oh_cols].sum()).mean()
median_non_low_exercise = (non_night_low_df.groupby('day')[exercise_oh_cols].sum()).median()
print('Mean Exercise Before Good Nights:\n' + str(mean_non_low_exercise))
print('Median Exercise Before Good Nights:\n' +  str(median_non_low_exercise))
```
