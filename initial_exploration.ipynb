### Analyze number of pod readings over time
- The count columns says how many times the pod recorded measurments in a day
- Data from this column can be examined to see when the pod was working well by plotting the count over time 

```python
def graph_labels(ax, xlabel, ylabel, title):
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    return ax
```

```python
import seaborn as sns
import numpy as np
%matplotlib inline

full_days_df = summary_df.loc[1:]

measurments_over_time = sns.lineplot(x=full_days_df.index, y=full_days_df['Count'])
# measurments_over_time.set_xticklabels(measurments_over_time.get_xticklabels(), rotation=90)
labelled_measurments = graph_labels(measurments_over_time, 'Days Ago (From 17/04/21)', 'Num Pod Measurments', 'Pod Measurements over Time')
```

### Examine the means and medians of in range, low, and high over the last 6 months
- Graph using pie charts since these are percentages

```python
in_range_means = summary_df['In Range'].mean()
low_means = summary_df['Low'].mean()
high_means = summary_df['High'].mean()
glucose_range_means = [in_range_means, low_means, high_means]
labels = ['In Range', 'Low', 'High']
print(glucose_range_means)
```

```python
import matplotlib.pyplot as plt

pie, ax = plt.subplots(figsize=[10, 6])
plt.title("Mean Percentage of Glucose Zones")
plt.pie(glucose_range_means, autopct='%1.1f%%',
        pctdistance=1.2, labeldistance=1.35)
plt.legend(labels=labels, loc='center left', bbox_to_anchor=(1.01, 0.5))
```
