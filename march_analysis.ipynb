```python
only_night_lows = df.loc[(df['mmol/L'] < 5) & ((df['hours_time'] > 23) | (df['hours_time'] < 8))].reset_index()
only_night_lows
```

```python
import numpy as np
import seaborn as sns
# draw a histogram to see lows by day of week - overshoot by 2 to reach the 31st
bounds = np.arange(1, 33)
hist, edges = np.histogram(only_night_lows['day'], bounds)
rel_freq = hist/float(hist.sum())
sns.set(rc={'figure.figsize': (10, 6)})
march_lows = sns.barplot(x=bounds[:-1], y=rel_freq)
march_lows.set_xlabel('Day')
march_lows.set_ylabel('% Lows')
march_lows.set_title('March Night Lows')
```

# Conduct a Statistical Test on Walking
- Walking seems to be normally distributed and differ between both categories
- The variables are independant
- Use a Levene test to check the variablility

```python
import scipy.stats as stats

levene_test = stats.levene(
    all_execise_df.loc[all_execise_df['night_bs'] == 'in_range', 'Walking'],
    all_execise_df.loc[all_execise_df['night_bs'] == 'low', 'Walking']
    )

print(levene_test)
```

- The p value is above the alpha cutoff of 0.05 so the variances can be considered equivalent - there is homogeneity of varience (HOV)
- All of the conditions for the t-test are met - the t test for walking can be conducted

```python
# population varience 
walking_t_test = stats.ttest_ind(
    all_execise_df.loc[all_execise_df['night_bs'] == 'in_range', 'Walking'],
    all_execise_df.loc[all_execise_df['night_bs'] == 'low', 'Walking']
)
walking_t_test
```
