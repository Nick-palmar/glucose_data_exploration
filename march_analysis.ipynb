# Conduct a Statistical Test on Walking
- Walking seems to be normally distributed (from violin plots above)
- The observations are independant of one another 
- Use a Levene test to check the variablility

```python
import scipy.stats as stats

levene_test = stats.levene(
    all_execise_df.loc[all_execise_df['night_bs'] == 'in_range', 'Walking'],
    all_execise_df.loc[all_execise_df['night_bs'] == 'low', 'Walking']
    )

print(levene_test)
```

- The p value is above the alpha cutoff of 0.05 so I can accept the Levene test null hypothesis and variances can be considered equivalent - there is homogeneity of varience (HOV)
- All of the conditions for the t-test are met 
- **Hypothesis**: Higher amounts of walking during the day lead to low blood sugars at night
- **Null Hypothesis**: There is no difference between the amount of walking for nights that are 1 or 2 days before lows and non-low nights

```python
# population varience 
walking_t_test = stats.ttest_ind(
    all_execise_df.loc[all_execise_df['night_bs'] == 'in_range', 'Walking'],
    all_execise_df.loc[all_execise_df['night_bs'] == 'low', 'Walking']
)
walking_t_test
```

# Outcomes
- **p=0.03** therefore, reject the null hypothesis: There is a difference betweem the amount of walking for low and non-low nights
- There is a statistically significant difference between the amount of walking my brother does and weather or not he will have a low blood sugar at night. From the data, it can be inferred that more walking implies that there is a greater chance of getting low at night
- To make better use of this difference, data can be cleaned and labelled on a per night basis
- A model can then be trained on this data to predict if he will get low based on exercise quantities. Given that walking has statistically significant results, the model should pick up on trends in the data
