### Analyze number of pod readings over time

- Data from this column can be examined to see when the pod was working well by plotting the count over time 

```python
def graph_labels(ax, xlabel, ylabel, title):
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    return ax
```

### Graph Blood sugar and exercise by month

```python
april_df = summary_df.loc[summary_df['month'] == 'Apr']
april_df = april_df.loc[::-1].reset_index(drop=True)
april_df.head()
```

```python
sns.set(rc={'figure.figsize': (12, 5)})
sns.set_style('dark')
apr_exercise_ax = sns.barplot(x=april_df.index, y=april_df['Total Exercise (mins)'])
apr_bs_ax = apr_exercise_ax.twinx()

sns.lineplot(x=april_df.index, y=april_df['Average (mmol/L)'], ax=apr_bs_ax, marker='o')
apr_bs_ax.set_ylim([0, 13])
```

```python
from typing import Tuple
def exercise_and_blood_sugar(month: str, exercise_ax: plt.Axes, in_range: Tuple[int] = (4, 10)):
    """ Graph exercise and blood sugars on the same axis for a specific month

    Args:
        month: The first 3 letters of the month
        exercise_ax: The axes to graph on; the initial exercise axis
        in_range: A tuple containing the minimum and maximum in range values
    
    Returns:
        A dual axis chart that shows amout of exercise with trends in bs for all days in the month
    """
    label_font = 30
    params = {
        'axes.labelsize': label_font,
   
        'xtick.labelsize': 20,
        'ytick.labelsize': 20,
        'axes.titlesize': 60
    }
    plt.rcParams.update(params)
    # sns.set_context("paper", rc={"font.size":80,"axes.titlesize":80,"axes.labelsize":50})  

    month_dict = {
        'Nov': 'November',
        'Dec': 'December',
        'Jan': 'January',
        'Feb': 'February',
        'Mar': 'March',
        'Apr': 'April'
    }

    df = summary_df.loc[summary_df['month'] == month]
    df = df.loc[::-1].reset_index(drop=True)

    # sns.set(rc={'figure.figsize': (12, 5)})
    sns.set_style('dark')
    sns.barplot(x=df.index, y=df['Total Exercise (mins)'], ax=exercise_ax)
    bs_ax = exercise_ax.twinx()

    sns.lineplot(x=df.index, y=df['Average (mmol/L)'], ax=bs_ax, marker='o', linewidth=6)
    bs_ax.set_ylim([0, 13])
    bs_ax.set_title(f'{month_dict[month]}: Exercise vs Blood Sugars')
    bs_ax.set_ylabel('Average Blood Sugar(mmol/L)', fontsize=label_font)
    exercise_ax.set_xlabel('Day in Month')

    # show in range blood sugars for clarity
    bs_ax.axhline(y=in_range[1], color='k', linewidth=2)
    bs_ax.axhline(y=in_range[0], color='k', linewidth=2)
    

    return exercise_ax
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12*6, 5*6))

all_months = ['Jan', 'Feb', 'Mar', 'Apr']
row_index = 0
col_index = 0 
for month in all_months:
    exercise_and_blood_sugar(month, ax[row_index, col_index])
    row_index += 1

    if row_index == 2:
        row_index = 0
        col_index += 1

# ax[2, 1] = sns.lineplot(x=[i for i in range(10)], y=[i for i in range(10)])
# ax[1, 1] = sns.lineplot(x=[i for i in range(10)], y=[i for i in range(10)])
```
